```python
import numpy as np

def avg_word_count(series):
    word_count = series.apply(lambda x: len(str(x).split()))
    return (word_count.mean(), word_count.std())

def get_ngrams(text, n):
    words = str(text).split()
    return set(tuple(words[i:i+n]) for i in range(len(words)-n+1))

def percent_novel_ngrams(df, n):
    abstracts_ngrams = df['abstract'].apply(lambda x: get_ngrams(x, n))
    annotations_ngrams = df['annotation'].apply(lambda x: get_ngrams(x, n))
    novel_counts = []
    total_counts = []
    for ann, abs_ in zip(annotations_ngrams, abstracts_ngrams):
        novel = ann - abs_
        novel_counts.append(len(novel))
        total_counts.append(len(ann))
    mean = np.mean([nc/tc if tc > 0 else 0 for nc, tc in zip(novel_counts, total_counts)])
    std = np.std([nc/tc if tc > 0 else 0 for nc, tc in zip(novel_counts, total_counts)])
    return (mean * 100, std * 100)

datasets = {'full': full, 'train': train, 'test': test}
results = {}

for name, df in datasets.items():
    results[name] = {
        'abstract_avg_words': avg_word_count(df['abstract']),
        'annotation_avg_words': avg_word_count(df['annotation']),
        'annotation_novel_ngrams_1': percent_novel_ngrams(df, 1),
        'annotation_novel_ngrams_2': percent_novel_ngrams(df, 2),
        'annotation_novel_ngrams_3': percent_novel_ngrams(df, 3),
    }

for name, stats in results.items():
    print(f"Dataset: {name}")
    print(f"  Avg. # of Words in abstract: {stats['abstract_avg_words'][0]:.2f}±{stats['abstract_avg_words'][1]:.2f}")
    print(f"  Avg. # of Words in annotation: {stats['annotation_avg_words'][0]:.2f}±{stats['annotation_avg_words'][1]:.2f}")
    print(f"  % novel 1-grams in annotation: {stats['annotation_novel_ngrams_1'][0]:.2f}±{stats['annotation_novel_ngrams_1'][1]:.2f}%")
    print(f"  % novel 2-grams in annotation: {stats['annotation_novel_ngrams_2'][0]:.2f}±{stats['annotation_novel_ngrams_2'][1]:.2f}%")
    print(f"  % novel 3-grams in annotation: {stats['annotation_novel_ngrams_3'][0]:.2f}±{stats['annotation_novel_ngrams_3'][1]:.2f}%\n")
```
